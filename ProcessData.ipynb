```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
PATH = Path('data').absolute()
```

# Processing Training Data

```python
train = pd.read_csv(PATH/'train.csv').drop('id', axis=1)
```

```python
train.head()
```

```python
is_valid = np.zeros(len(train), int)
is_valid[:] = False
is_valid[np.random.choice(len(train), size=int(len(train)*0.2), replace=False)] = True

train['is_valid'] = is_valid


for column in train.columns[1:-1]:
  train[column] = train[column].apply(lambda x: column if x == 1 else '')


def concat_labels(labels):
  s = ''
  for label in labels:
    if label != '':
      s = s + label + '|'
  return s[:-1]

train['target'] = train.iloc[:, 1:-1:].apply(lambda x : concat_labels(x), axis=1)


train = train.drop(['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'], axis=1)
train = train.loc[:, ['comment_text', 'target', 'is_valid']]
train
```

```python
train.to_csv(PATH/'train_processed.csv', index=False)
```
